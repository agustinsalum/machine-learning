## 1- Ingrese el DATA SET

```python
# Importar las bibliotecas necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Testeo, entrenamiento, regresion, metricas y codificacion de datos
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, roc_curve, roc_auc_score
from sklearn.feature_extraction.text import TfidfVectorizer
import seaborn as sns
```

## 4- Aplicar PROCESO para DEFINIR tu X e Y para APLICAR REGRESION LINEAL Y LOGISTICA segun el DATA SET a las variables que consideres adecuadas. RECUERDA ESTANDARIZAR LAS VARIABLES PARA CORRELACIÓN LOGÍSTICA.

La regresión lineal se utiliza para problemas de regresión numérica, no para clasificación binaria (spam vs. no spam). Dado que el problema implica clasificación, la regresión lineal podría no ser la técnica más apropiada para este caso. Debemos enfocarnos en técnicas de clasificación, como la regresión logística, en lugar de regresión lineal para este problema específico.

La variable dependiente (Y) representa las clases que intentamos predecir (Category). La variable independiente (X) representa las características o atributos que utilizamos para hacer estas predicciones.

```python
# Cambiamos los valores de categoria a 0 y 1
df['Category'] = df['Category'].map({'ham':0,'spam':1})

# Definir X e Y
x = df['Message'] 
y = df['Category']

# Verificamos
print(df.head(10))

```

Dividir los datos de entrenamiento y testeo.Asignar el 80% de los datos al conjunto de entrenamiento (x_train y y_train). Asignar el 20% restante al conjunto de prueba (x_test y y_test).

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2) 
```

Crear un vectorizador TF-IDF para convertir el texto en características numéricas. Se ajusta con 5000 para que sea mas exacta

```python
vectorizador = TfidfVectorizer(max_features=5000)

# Modificamos los valores de entrenamiento y testeo por la conversion vectorizada numerica del texto
x_train_tfid = vectorizador.fit_transform(x_train)
x_test_tfid = vectorizador.transform(x_test)
```

Entrenamos nuestro modelo de regresion logistica y realizamos la prediccion

```python
# Entrenar un modelo de regresión logística
logistic_regression_model = LogisticRegression()
logistic_regression_model.fit(x_train_tfid, y_train)

# Realizar predicciones en el conjunto de prueba
y_pred = logistic_regression_model.predict(x_test_tfid)
```

y_pred contiene las predicciones del modelo de regresión logística en el conjunto de prueba. Para evaluar el rendimiento del modelo, podemos utilizar métricas como precisión, recall, F1-score y la matriz de confusión.

```python
#Calculamos la probabilidad de predicción para el conjunto de prueba

y_prob = logistic_regression_model.predict_proba(x_test_tfid)[:, 1]
print(y_prob)
```

## 5- Aplicar alguna METRICAS de EVALUACION a los modelos anteriores y justificar la elección de dicha métrica.

Para un problema de clasificación de spam, se suele dar más importancia a la exhaustividad (recall) y a la precisión.

Un alto recall significa que el modelo es capaz de detectar la mayoría de los correos electrónicos de spam presentes en el conjunto de datos. Esto es crucial para garantizar que los correos electrónicos de spam no sean pasados por alto y terminen en la bandeja de entrada del usuario. En el contexto de spam, perder un correo electrónico legítimo no es tan grave como permitir que un correo electrónico de spam llegue a la bandeja de entrada del usuario. Un alto recall garantiza que la mayoría de los correos electrónicos de spam sean detectados, minimizando así los falsos negativos.

Una alta precisión significa que los correos electrónicos marcados como spam son, en su mayoría, spam genuino y no correos electrónicos legítimos. En el contexto de spam, marcar un correo electrónico legítimo como spam (falso positivo) puede ser una experiencia frustrante para los usuarios. Si el modelo tiene una alta precisión, los usuarios confiarán en que los correos electrónicos marcados como spam realmente son spam, lo que mejora la experiencia del usuario y evita que correos electrónicos importantes sean etiquetados incorrectamente.

```python

precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
print(f'Precisión: {precision:.2f}')
print(f'Exhaustividad: {recall:.2f}')
```
